```python
from langchain.chat_models import  ChatOpenAI
from langchain  import PromptTemplate , LLMChain
from langchain.prompts.chat import (
    ChatPromptTemplate ,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate, 
)
from langchain.schema import (
    AIMessage,
    HumanMessage,
    SystemMessage,
)
```

```python
llm = ChatOpenAI(model_name="gpt-4",temperature=0.25,openai_api_key=api_key_gpt,max_tokens=1000)
```

```python
from backcall.backcall import sys
template ="You are a helpful expert on intellectual property laws in Morocco. Please answer the question in detail in the same language as the question was asked. {lawdocs}"
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template="{question}"
human_messa_prompt = HumanMessagePromptTemplate.from_template(human_template)
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt,human_messa_prompt])
chain = LLMChain(llm=llm,prompt=chat_prompt)
```

```python
quest="ch7al men mara ne9eder nejadede lmarka deyali"
docsprompt=analysedocpdf.similarity_search(quest,include_metadata=True)
```
